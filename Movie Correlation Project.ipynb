```python
#importing libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)


# Now we need to read in the data
df = pd.read_csv(r'F:\Portfolio Project\pythion project\movies.csv')
```

```python
# Now lets look at the data
df.head()
```

```python
# DATA CLEANING

#Lets see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Lets look at the Data types of the columns

df.dtypes
```

```python
# Are there any Outliers?

df.boxplot(column=['gross'])
```

```python
#drop any duplicates

df.drop_duplicates()
```

```python
# let sort the data by gross from highest to lowest

df.sort_values(by = ['gross'], inplace=False ,ascending=False)
```

```python
#scatter plot Budget vs Gross earnings

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross earnings')
plt.xlabel('Gross earnings')
plt.ylabel('Budget')
plt.show()
```

```python
#lets find the correlation
# ploting the Budget vs Gross using seborn  

sns.regplot(x='gross', y='budget', data=df)
```

```python
sns.regplot(x='gross', y='budget', data=df, line_kws={'color':'blue'})
```

```python
#Alright we now know that they are correlatd, but lets find how much they are correlated.

df.corr()
```

```python
#let try it out with different methods

df.corr(method='pearson')
```

```python
df.corr(method='kendall')
```

```python

df.corr(method='spearman')
```

```python
#Now lets Visualize since its difficult to understand by the numbers

correlation_matrix = df.corr("pearson")

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes # this is where the random numerization happens
        
df_numerized 
```

```python
correlation_matrix = df_numerized.corr("pearson")

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# we can see it here not the gross and budget correlated but also the votes are coorelated with the gross.
```
